```python

new_columns = raw_data.columns.tolist()
new_columns.insert(2, 'content')
data = raw_data.reindex(columns=new_columns)
print(data.loc[:, 'title'].isnull().sum().sum())
data.loc[:,'content'] = data.loc[:, 'title'] + ' ' + data.loc[:, 'selftext']
data.fillna(method = 'ffill', axis = 1, inplace=True)
data.drop(['title', 'selftext'], axis = 1, inplace=True)
# data = pd.concat([data,label],axis=1)
data
```

```python
import sys
import nltk
nltk.download('stopwords')
nltk.download('punkt')
import sklearn
import pandas
import numpy

print('Python: {}'.format(sys.version))
print('NLTK: {}'.format(nltk.__version__))
print('Scikit-learn: {}'.format(sklearn.__version__))
print('Pandas: {}'.format(pandas.__version__))
print('Numpy: {}'.format(numpy.__version__))
```

```python
text_messages = data['content']
# print(text_messages)
# Replace numbers with 'numbr'
processed = text_messages.str.replace(r'\d+(\.\d+)?', 'numbr')

# Remove punctuation
processed = text_messages.str.replace(r'[^\w\d\s]', ' ')

# Replace whitespace between terms with a single space
processed = processed.str.replace(r'\s+', ' ')

# Remove leading and trailing whitespace
processed = processed.str.replace(r'^\s+|\s+?$', '')
```

```python
# change words to lower case - Hello, HELLO, hello are all the same word
processed = processed.str.lower()
processed
```

```python
from nltk.corpus import stopwords

# remove stop words from text messages

stop_words = set(stopwords.words('english'))

processed = processed.apply(lambda x: ' '.join(
    term for term in x.split() if term not in stop_words))
```

```python
# Remove word stems using a Porter stemmer
ps = nltk.PorterStemmer()

processed = processed.apply(lambda x: ' '.join(
    ps.stem(term) for term in x.split()))
processed
```

```python
from nltk.tokenize import word_tokenize

# create bag-of-words
all_words = []

for message in processed:
    words = word_tokenize(message)
    for w in words:
        all_words.append(w)

all_words = nltk.FreqDist(all_words)
```

```python
print('Number of words: {}'.format(len(all_words)))
print('Most common words: {}'.format(all_words.most_common(15)))
```

```python
word_features = list(all_words.keys())[:2448]
```

```python
attempts_label = label['Outcomes - Self-Harm/Suicide Attempts.1']
planning_label = label['Outcomes - Ideation/Planning.1']
print(attempts_label, planning_label)
```

```python
train_content = processed
train_attempts_label = attempts_label
train_planning_label = planning_label

def find_features(message):
    words = word_tokenize(message)
    features = {}
    for word in word_features:
        features[word] = (word in words)

    return features
```

Planning part adjustment

```python
# choose the classifier from ['SVC', 'KNN', 'Logistic', 'DecisionTree', 'SGD']
classifier = 'SVC'

planning_pkg = zip(train_content, train_planning_label)

featuresets = [(find_features(text), label) for (text, label) in planning_pkg]

from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import model_selection
from nltk.classify.scikitlearn import SklearnClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.feature_extraction import DictVectorizer
import numpy as np


#training, testing = model_selection.train_test_split(featuresets, test_size = 0.2, random_state= 0)
vectorizer = DictVectorizer()

X_train, y_train = list(zip(*featuresets))
X_train = vectorizer.fit_transform(X_train)

X_test, y_test = list(zip(*featuresets))
X_test = vectorizer.transform(X_test)

if classifier == 'SGD':
    param_test = {'alpha':np.arange(0.1, 5, 0.2)}
    estimator = SGDClassifier(loss = 'hinge', random_state=0)
elif classifier == 'SVC':
    param_test = {'C':np.arange(0.01, 2, 0.01)}
    estimator = SVC(kernel = 'linear',random_state =0)
elif classifier == 'KNN':
    param_test = {'n_neighbors':np.arange(5, 100, 5), 'leaf_size':np.arange(1, 10, 1)}
    estimator = KNeighborsClassifier(leaf_size = 1, weights = 'distance', n_neighbors = 41)
elif classifier == 'Logistic':
    param_test = {'C':np.arange(0.1, 2, 0.1)}
    estimator = LogisticRegression(class_weight='balanced', solver='liblinear', penalty='l1')
elif classifier == 'DecisionTree':
    param_test = {'min_samples_leaf':np.arange(1, 15, 1), 'min_samples_split':np.arange(1, 10, 1)}
    estimator = DecisionTreeClassifier()
else:
    print('classifier error')

gsearch1 = GridSearchCV(estimator =  estimator,
                       param_grid = param_test, scoring='roc_auc',cv=10)

# Define models to train
gsearch1.fit(X_train, y_train)
print(gsearch1.best_params_, gsearch1.best_score_)
```
