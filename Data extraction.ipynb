```python
#Libraries in use
import numpy as np
import pandas as pd
import openpyxl
```

```python
# click the URL and paste the obtained code into the bar and click the Enter key
from google.colab import drive
drive.mount('/content/drive')
```

```python
#Users enter the parmeters of there search 
startMonth=int(input("Please enter the number of month you would like to start at: "))
startYear=int(input("Please enter the year you would like to start at: "))
numberOfMonths=int(input("How many months would you like? "))
numberOfRows=int(input("How many rows of data would you like?"))
outputFileName = input("PLease enter the name of the output file: ") + ".xlsx"
```

```python
#Intializing variables
Months=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
root = '/content/drive/MyDrive/ActiveLearningStigma/Raw Data'
fileName = "/" + Months[startMonth-1] + str(startYear) +"_Submissions.xlsx"
index = 0 
finalData = pd.DataFrame()
```

```python
#Cycling through the selected months and pulling data to combine and be saved
while index != numberOfMonths:
  data = pd.read_excel(root+fileName)
  specificRows= data.iloc[:numberOfRows]
  finalData=finalData.append(specificRows)
  if Months[startMonth-1] == 'December':
    startYear += 1
    startMonth = -1
  fileName = "/" + Months[startMonth+1] + str(startYear) +"_Submissions.xlsx"
  index+=1
```

```python
#Conveting to excel file and saving to drive 
finalData.to_excel('/content/drive/MyDrive/ActiveLearningStigma/'+outputFileName)
```
